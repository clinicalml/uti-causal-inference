```python
# name_map = OrderedDict()
# # Create a dictionary for mapping short names with proper names
# name_map.update({
#     "t_bin": "All relevant revisits",
#     "t_uti_bin": "UTI",
#     "t_neph_bin": "Pyelonephritis",
#     "t_sepsis_bin": "Sepsis",
#     "t_i_bin": "All relevant inpatient revisits",
#     "t_i_uti_bin": "2_UTI",
#     "t_i_neph_bin": "2_Pyelonephritis",
#     "t_i_sepsis_bin": "2_Sepsis",
#     "AE_any":"All adverse events",
#     "AE_c_diff":"C. difficile infection",
#     "AE_skin":"Skin",
#     "AE_GI":"Gastrointestinal",
#     "AE_AKI":"Acute kidney injury",
#     "AE_other":"Other",
# }
# )

name_map = OrderedDict()
# Create a dictionary for mapping short names with proper names
# name_map.update({
name_map.update({
    "t_bin": "All relevant revisits",
    "t_uti_bin": "UTI",
    "t_neph_bin": "Pyelonephritis",
    "t_sepsis_bin": "Sepsis",
    "t_i_bin": "All relevant inpatient revisits",
    "t_i_uti_bin": "2_UTI",
    "t_i_neph_bin": "2_Pyelonephritis",
    "t_i_sepsis_bin": "2_Sepsis",
    "AE_any":"All adverse events",
    "AE_c_diff":"C. difficile infection",
    "AE_skin":"Skin",
    "AE_GI":"Gastrointestinal",
    "AE_AKI":"Acute kidney injury",
    "AE_other":"Other",
    "fibro_1_mon_outcome":"One month Fibrocystic Disease",
    "fibro_3_mon_outcome":"Three month Fibrocystic Disease",
    "fibro_6_mon_outcome":"Six month Fibrocystic Disease",
    "hernia_1_mon_outcome":"One Month Hernia",
    "hernia_3_mon_outcome":"Three Month Hernia",
    "hernia_6_mon_outcome":"Six Month Hernia",
    "fracture_1_mon_outcome":"One Month Fracture",
    "fracture_3_mon_outcome":"Three Month Fracture",
    "fracture_6_mon_outcome":"Six Month Fracture"
}
)

export_csv_name_map =  name_map

export_csv_name_map.update({"t_i_uti_bin": "UTI (inpatient)",
    "t_i_neph_bin": "Pyelonephritis (inpatient)",
    "t_i_sepsis_bin": "Sepsis (inpatient)",
                           })

def round_with_padding(value, round_digits):
    return format(round(value,round_digits), "."+str(round_digits)+"f")

# print(name_map)
# Find all the relevant data using regex
def forest_plot_format(text,display=False):
#     matches = re.findall(r"ATE for ((?:t_|AE_)[A-Z|a-z_]+) : (-?\d+\.?\d*(?:[eE]-?\d+)?)+\n95% C\.I\.: \( (-?\d+\.?\d*(?:[eE]-?\d+)?) ,\s+(-?\d+\.?\d*(?:[eE]-?\d+)?) \)", text)# Transform the raw data into the desired list format
#     print(matches,end="\n\n")
    #     matches = re.findall(
#         r"ATE for ((?:t_|AE_|fibro|hernia|fracture)[A-Z|a-z_]+(?:1|3|6)_mon_outcome) : (-?\d+\.?\d*(?:[eE]-?\d+)?)+\n95% C\.I\.: \( (-?\d+\.?\d*(?:[eE]-?\d+)?) ,\s+(-?\d+\.?\d*(?:[eE]-?\d+)?) \)",
#         text
#     )
    matches = re.findall(
        r"ATE for ((?:t_|AE_|fibro|hernia|fracture)[A-Z|a-z_]+(?:(?:1|3|6)_mon_outcome)?) : (-?\d+\.?\d*(?:[eE]-?\d+)?)+\n95% C\.I\.: \( (-?\d+\.?\d*(?:[eE]-?\d+)?) ,\s+(-?\d+\.?\d*(?:[eE]-?\d+)?) \)",
        text
    )
#     print(matches)
    data = [(name_map[item[0]], '{:.2f}'.format(round(-float(item[1]) * 100,2)), ('{:.2f}'.format(round(-float(item[3]) * 100,2)), '{:.2f}'.format(round(-float(item[2]) * 100,2)))) for item in matches]
    # Print the result
    formatted_string = "["
    if display:
        print(" = [")
    for item in name_map.values():
        for datum in data:
            if datum[0].startswith(item):
                if display:
                    print(datum,end=",\n")
                formatted_string += str(datum)
                formatted_string += ","
    if display:
        print("]")
    formatted_string += "]"
    return formatted_string
    

def format_sign(metric_string):
    if '-' in metric_string:
        metric_string = metric_string.replace('-', 'âˆ’')
    else:
        metric_string = '   ' + metric_string
    return metric_string
    
# Function to plot forest plot
def create_forest_plot_secondline(data, x_label, y_label,data_size,figsize=(5.5,4.5),plt_format='pdf',save_name="second_line_forest_plot"):
    fig, ax = plt.subplots(figsize=figsize)
    data_min,data_max=data_size
    for i, row in enumerate(data):
        name, point, ci = row
        ax.plot([ci[0], ci[1]], [i, i], color='lightgray', linewidth=2)
        ax.plot([point], [i], marker='o', markersize=3, color='black')
        ax.vlines(ci[0], i - 0.2, i + 0.2, color='lightgray', linewidth=1) # Add vertical line on the left of CI interval
        ax.vlines(ci[1], i - 0.2, i + 0.2, color='lightgray', linewidth=1) # Add vertical line on the right of CI interval
        x_pos = -9.0  # Default padding for y labels (modify to your preference)
        if name in ["All relevant revisits","All relevant inpatient revisits","All adverse events"]:  # Modify padding only for this label
            x_pos -= 0.10

        ax.text(x_pos, i, name, fontsize=10, verticalalignment='center', horizontalalignment='left')

    #add vertical line at origin
    ax.axvline(0, color='lightgray', linestyle='-')
    min_val = min([d[2][0] for d in data])
    max_val = max([d[2][1] for d in data])
    pad = 0.4 * (max_val - min_val)
    
    for i, row in enumerate(data):
        name, point, ci = row
        # Display data point value and the CI interval on the right side of the plot (outside the plot area)
        point_string = format_sign(f"{point:.2f}")
        lci = format_sign(f"{ci[0]:.2f}")
        rci = format_sign(f"{ci[1]:.2f}")
        
        metric_string =  point_string + " (" + lci + ", " + rci + ")"
        ax.text(max_val + pad, i, metric_string, fontsize=10, horizontalalignment='left', verticalalignment='center')    
    # Set labels
    ax.set_xlabel(x_label)
#     ax.set_ylabel(y_label)
    
    y_labels = [d[0] for d in data]
    y_ticks = range(len(y_labels))
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(y_labels)


    min_tick = min([d[2][0] for d in data])  # Find minimum CI value
    max_tick = max([d[2][1] for d in data])  # Find maximum CI value
    x_ticks = np.arange(data_min, data_max, 0.5)
    ax.set_xticks(x_ticks)
    #use unicode dash and x_ticks then rotate 45
    ax.set_xticklabels([f"{x:g}".replace('-', 'âˆ’') for x in x_ticks],fontsize=10)
    # Adjust plot dimensions
    plt.subplots_adjust(left=0.4)
    ax = plt.gca()
    ax.invert_yaxis()
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.yticks([])
    if plt_format=='png':
        plt.savefig(f"{save_name}.png",dpi=200,bbox_inches='tight')
    elif plt_format =='pdf':
        plt.savefig(f"{save_name}.pdf",dpi=200,format="pdf", bbox_inches="tight")
    plt.show()
    
def create_forest_plot_alternatives(data, x_label, y_label,data_size,figsize=(5.5,4.5),plt_format='png',save_name="alternatives_forest_plot"):
    fig, ax = plt.subplots(figsize=figsize)
    data_min,data_max=data_size
    for i, row in enumerate(data):
        name, point, ci = row
        ax.plot([ci[0], ci[1]], [i, i], color='lightgray', linewidth=2)
        ax.plot([point], [i], marker='o', markersize=3, color='black')
        ax.vlines(ci[0], i - 0.2, i + 0.2, color='lightgray', linewidth=1) # Add vertical line on the left of CI interval
        ax.vlines(ci[1], i - 0.2, i + 0.2, color='lightgray', linewidth=1) # Add vertical line on the right of CI interval
        x_pos = -17.0  # Default padding for y labels (modify to your preference)
        if name in ["All relevant revisits","All relevant inpatient revisits","All adverse events"]:  # Modify padding only for this label
            x_pos -= 0.15

        ax.text(x_pos, i, name, fontsize=10, verticalalignment='center', horizontalalignment='left')

    #add vertical line at origin
    ax.axvline(0, color='lightgray', linestyle='-')
    min_val = min([d[2][0] for d in data])
    max_val = max([d[2][1] for d in data])
    pad = 0.15 * (max_val - min_val)
    
    for i, row in enumerate(data):
        name, point, ci = row
        # Display data point value and the CI interval on the right side of the plot (outside the plot area)
        point_string = format_sign(f"{point:.2f}")
        lci = format_sign(f"{ci[0]:.2f}")
        rci = format_sign(f"{ci[1]:.2f}")

        metric_string =  point_string + " (" + lci + ", " + rci + ")"
        
        ax.text(max_val + pad, i, metric_string, fontsize=10, horizontalalignment='left', verticalalignment='center')    
    # Set labels
    ax.set_xlabel(x_label)
#     ax.set_ylabel(y_label)
    
    y_labels = [d[0] for d in data]
    y_ticks = range(len(y_labels))
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(y_labels)


    min_tick = min([d[2][0] for d in data])  # Find minimum CI value
    max_tick = max([d[2][1] for d in data])  # Find maximum CI value
    x_ticks = np.arange(data_min, data_max, 0.5)
    ax.set_xticks(x_ticks)
    #use unicode dash and x_ticks then rotate 45
    ax.set_xticklabels([f"{x:g}".replace('-', 'âˆ’') for x in x_ticks], fontsize=10)
    # Adjust plot dimensions
    plt.subplots_adjust(left=0.4)
    ax = plt.gca()
    ax.invert_yaxis()
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.yticks([])
    if plt_format=='png':
        plt.savefig(f"{save_name}.png",dpi=200,bbox_inches='tight')
    elif plt_format =='pdf':
        plt.savefig(f"{save_name}.pdf",dpi=200,format="pdf", bbox_inches="tight")
    plt.show()

def remove_quotes(data):
    cleaned_data = []
    for entry in data:
        name = entry[0]
        value = ast.literal_eval(entry[1])
        lower_bound = ast.literal_eval(entry[2][0])
        upper_bound = ast.literal_eval(entry[2][1])
        cleaned_data.append((name, value, (lower_bound, upper_bound)))
    return cleaned_data

def dataframe_to_formatted_string(df):
    # Start with the header
    string_result = "cohort_features\n"
    
    # Iterate over each row of the dataframe
    for index, row in df.iterrows():
        # Extract row values
        outcome = row['Outcome']
        e_y1 = row['E[Y_1]']
        e_y0 = row['E[Y_0]']
        ate = row['ATE']
        ci_lower, ci_upper = row['95% C.I.']
        
        # Format the string for the current row
        row_string = f"ATE for {outcome} : {ate:.6f}\n"
        row_string += f"95% C.I.: ( {ci_lower:.6f} , {ci_upper:.6f} )\n"
        row_string += f"EY1: {e_y1:.6f}\n"
        row_string += f"EY0: {e_y0:.6f}\n"
        
        # Add the current row string to the result
        string_result += row_string
    
    return string_result

def remove_format_characters(lst):
    updated_metrics = []
    for tup in lst:
        if '2_' in tup[0]:

            new_tup = list(tup)

            new_tup[0] = tup[0].replace('2_','')
            tup = tuple(new_tup)

            
        updated_metrics.append(tup)
        
    return updated_metrics

def format_characters(lst):
    updated_metrics = []
    for tup in lst:
        new_tup = list(tup)
        if '2_' in tup[0]:
            new_tup[0] = tup[0].replace('2_','')
        if '-' not in tup[1]:
            new_tup[1] = ' ' + tup[1]
            
        tup = tuple(new_tup)
            
        updated_metrics.append(tup)
        
    return updated_metrics


def convert_nc_format(data):
    time_map = {'one month': 'One month',
               'three month': 'Three months',
               'six month': 'Six months'}
    def split_time_cond(time_condition,time_map):
        for time in time_map.keys():
            if time in time_condition.lower():
                condition_name = time_condition.split('onth')[1].lstrip()
                found_time = time_condition.replace(condition_name,'').rstrip().lower()
        return found_time, condition_name 
    
    time_cond_dict = defaultdict(dict)

    #get the condition and the name 
    for time_condition,ate,ci in data:
        time,condition = split_time_cond(time_condition,time_map)
        time_cond_dict[time][condition] = (ate,ci)
    
    formatted_list = []
    
    #append to the output
    for time in time_cond_dict.keys():
        time_header = (time_map[time],None,None)
        formatted_list.append(time_header)
        
        conditions = time_cond_dict[time].keys()

        for condition in conditions:
            ate, ci = time_cond_dict[time][condition][0], time_cond_dict[time][condition][1]
            formatted_list.append((condition,ate,ci))
    
    return formatted_list
```

```python
def convert_negative_control_format(omop_v1):
    # Create a dictionary to map the time periods to the corresponding string
    time_period_map = {
        'one month': 'One month',
        'three month': 'Three months',
        'six month': 'Six months'
    }
    
    # Initialize an empty list to hold the converted format
    text_omop_v2 = []
    
    # Loop over the input list and process each entry
    for entry in omop_v1:
        # Split the string to extract the time period and condition
        parts = entry[0].split(' ')
        time_period = ' '.join(parts[:-2]).lower()  # Convert to lowercase
        condition = parts[-2] + ' ' + parts[-1]
        
        # Map the time period to the desired format and add to the list if not already present
        time_period_formatted = time_period_map.get(time_period, None)
        if time_period_formatted and time_period_formatted not in (item[0] for item in text_omop_v2):
            text_omop_v2.append((time_period_formatted, None, None))
        
        # Add the condition with the corresponding values
        text_omop_v2.append((condition, entry[1], entry[2]))
    
    return text_omop_v2

# The test_omop_v1 list from your example

def convert_negative_control_format_2(omop_v1):
    # Define a function to normalize the time period strings
    def normalize_time_period(time_str):
        time_str = time_str.lower().strip()
        if 'one' in time_str:
            return 'one month'
        elif 'three' in time_str:
            return 'three months'
        elif 'six' in time_str:
            return 'six months'
        return time_str

    # Initialize an empty list to hold the converted format
    text_omop_v2 = []
    
    # Initialize an empty dictionary to keep track of the time periods
    time_periods_seen = {}

    # Loop over the input list and process each entry
    for entry in omop_v1:
        # Extract the time period and condition from the entry
        parts = entry[0].split(' ')
        time_period = ' '.join(parts[:-2])
        condition = parts[-2] + ' ' + parts[-1]

        # Normalize the time period
        normalized_time_period = normalize_time_period(time_period)

        # Check if the normalized time period has been seen before
        if normalized_time_period not in time_periods_seen:
            # If not, add the time period entry and mark it as seen
            text_omop_v2.append((normalized_time_period.title(), None, None))
            time_periods_seen[normalized_time_period] = True

        # Add the condition with the corresponding values
        text_omop_v2.append((condition, entry[1], entry[2]))

    return text_omop_v2


def convert_negative_control_format_3(data):
    """
    Transforms data from the first version format to the second version format.

    Args:
        data: A list of tuples representing the data in the first version format.

    Returns:
        A list of tuples representing the data in the second version format.
    """

    transformed_data = []
    current_month = None

    for condition, _, values in data:
        month, condition_name = condition.rsplit(" ", 1)
        month = month.lower()  # Apply lower() to the month string

        if month != current_month:
            transformed_data.append((month, None, None))
            current_month = month

        transformed_data.append((condition_name, *values))

    return transformed_data


```

```python
import matplotlib.pyplot as plt
import numpy as np

def create_forest_plot_secondline_2(data, x_label, y_label, data_size,pad_type, reduce_xticks=False,pad_num=0.5,pad_params=(None, None), figsize=(5.5,4.5), plt_format='pdf', save_name="second_line_forest_plot"):
    fig, ax = plt.subplots(figsize=figsize)
    data_min, data_max = data_size

    if pad_type == 'second_line':
        x_pos_default,x_pos_dif = -9,.1
    elif pad_type == 'alternatives':
        x_pos_default,x_pos_dif = -17,.15
    else:
        assert any(param is not None for param in pad_params), "pad params of type 'Tuple' or pad_type of type 'str' not provided"
        x_pos_default,x_pos_dif = pad_params
    for i, row in enumerate(data):
        name, point, ci = row
        if point is not None and ci is not None:
            ax.plot([ci[0], ci[1]], [i, i], color='lightgray', linewidth=2)
            ax.plot([point], [i], marker='o', markersize=3, color='black')
            ax.vlines(ci[0], i - 0.2, i + 0.2, color='lightgray', linewidth=1)
            ax.vlines(ci[1], i - 0.2, i + 0.2, color='lightgray', linewidth=1)

        x_pos = x_pos_default  # Default padding for y labels (modify to your preference)
        if name in ["All relevant revisits", "All relevant inpatient revisits", "All adverse events","One month","Three months","Six months"]:
            x_pos -= x_pos_dif
        ax.text(x_pos, i, name, fontsize=10, verticalalignment='center', horizontalalignment='left')
    
    ax.axvline(0, color='lightgray', linestyle='-')
    
    ax.set_xlabel(x_label)
    y_labels = [d[0] for d in data]
    y_ticks = range(len(y_labels))
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(y_labels)

    x_ticks = np.arange(data_min, data_max, 0.5)
    ax.set_xticks(x_ticks)
    ax.set_xticklabels([format_sign(f"{x:g}") for x in x_ticks], fontsize=10)

    plt.subplots_adjust(left=0.4)
    ax.invert_yaxis()
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.yticks([])

    # Determine the right edge of the plot for placing the text
    right_edge = ax.get_xlim()[1]
    text_pad = pad_num * right_edge  # Set a padding relative to the right edge
    
    for i, row in enumerate(data):
        name, point, ci = row
        if point is not None and ci is not None:
            # Display data point value and the CI interval on the right side of the plot (outside the plot area)
            point_string = format_sign(f"{point:.2f}")
            lci = format_sign(f"{ci[0]:.2f}")
            rci = format_sign(f"{ci[1]:.2f}")
                
            metric_string = point_string + " (" + lci + ", " + rci + ")"
            ax.text(right_edge + text_pad, i, metric_string, fontsize=10, horizontalalignment='left', verticalalignment='center')

    if reduce_xticks:
        for label in ax.xaxis.get_ticklabels()[1::2]:
            label.set_visible(False)
            
    if plt_format == 'png':
        plt.savefig(f"{save_name}.png", dpi=200, bbox_inches='tight')
    elif plt_format == 'pdf':
        plt.savefig(f"{save_name}.pdf", dpi=200, format="pdf", bbox_inches="tight")
    plt.show()
    
```
